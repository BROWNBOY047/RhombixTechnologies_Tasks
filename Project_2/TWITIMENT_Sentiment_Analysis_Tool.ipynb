# **1 __ Loading & Checking the IMDB Dataset**

```python
from datasets import load_dataset

# Load IMDB dataset
imdb = load_dataset("imdb")

# Show structure
print(imdb)

# Show sample
print("\nSample example:")
print(imdb['train'][0])
```

# **2 __ Loading &  Mapping GoEmotions Emotions to Sentiments**

```python
from datasets import load_dataset

# Load the GoEmotions dataset
goemotions = load_dataset("go_emotions")

# Get label names
label_names = goemotions['train'].features['labels'].feature.names

# Define emotion to sentiment mapping
positive_emotions = {'admiration', 'amusement', 'approval', 'curiosity', 'excitement', 'gratitude', 'joy', 'love', 'optimism', 'pride', 'relief'}
negative_emotions = {'anger', 'annoyance', 'disgust', 'disapproval', 'embarrassment', 'fear', 'grief', 'nervousness', 'disappointment', 'remorse', 'sadness'}
neutral_emotions = {'neutral'}

# Map emotions to sentiment labels
def map_emotions(example):
    emotions = [label_names[l] for l in example['labels']]
    sentiment = None
    for emo in emotions:
        if emo in positive_emotions:
            sentiment = 2  # Positive
            break
        elif emo in negative_emotions:
            sentiment = 0  # Negative
            break
        elif emo in neutral_emotions:
            sentiment = 1  # Neutral
            break
    return {'sentiment_label': sentiment}

# Apply mapping
goemotions_labeled = goemotions['train'].map(map_emotions)

# Filter out entries without a sentiment label
goemotions_labeled = goemotions_labeled.filter(lambda x: x['sentiment_label'] is not None)
```

# **3 __ Tokenizing the GoEmotions dataset**

```python
from transformers import AutoTokenizer

# Load the tokenizer used in your model
tokenizer = AutoTokenizer.from_pretrained("cardiffnlp/twitter-roberta-base-sentiment")

# Define a function to tokenize the text
def tokenize_function(example):
    return tokenizer(example["text"],max_length=64, truncation=True, padding="max_length")

# Apply tokenization
tokenized_goemotions = goemotions_labeled.map(tokenize_function, batched=True)

# Rename 'sentiment_label' to 'labels' for training compatibility
if "labels" in tokenized_goemotions.column_names:
    tokenized_goemotions = tokenized_goemotions.remove_columns("labels")
if "sentiment_label" in tokenized_goemotions.column_names:
    tokenized_goemotions = tokenized_goemotions.rename_column("sentiment_label", "labels")

# Set the format for TensorFlow (or PyTorch)
tokenized_goemotions.set_format("tensorflow")  # or use "torch" if you're using PyTorch
```

# **4 __ Tokenizing the IMDB dataset**

```python
from transformers import AutoTokenizer

# Load the tokenizer
tokenizer = AutoTokenizer.from_pretrained("cardiffnlp/twitter-roberta-base-sentiment")

# Define a function for tokenization
def tokenize_function(example):
    return tokenizer(example["text"],max_length=64, padding="max_length", truncation=True)

# Tokenize the IMDB dataset
tokenized_imdb = imdb.map(tokenize_function, batched=True)
```

# **5 __ Mapping IMDB Labels to Match GoEmotions Labels**

```python
# Check label names (optional, just for your understanding)
print(imdb["train"].features["label"])

# Rename label column to 'sentiment_label' for consistency
def rename_label(example):
    return {"sentiment_label": example["label"]}

# Apply to tokenized IMDB
tokenized_imdb = tokenized_imdb.map(rename_label)
```

# **6 __ Aligning and Combining Both Datasets**

```python
from datasets import concatenate_datasets

# Use only the training split from IMDB
tokenized_imdb_train = tokenized_imdb["train"]

# Combine the tokenized datasets
combined_dataset = concatenate_datasets([tokenized_goemotions, tokenized_imdb_train])

# Shuffle the combined dataset
combined_dataset = combined_dataset.shuffle(seed=42)

# Set format for TensorFlow
combined_dataset.set_format("tensorflow")
```

# **7 __ Converting the Combined Dataset into TensorFlow-Compatible Tensors for Training**

```python
from datasets import Dataset, concatenate_datasets
from transformers import AutoTokenizer, DataCollatorWithPadding
import tensorflow as tf

# Fix label columns
if "sentiment_label" in tokenized_goemotions.column_names:
    tokenized_goemotions = tokenized_goemotions.rename_column("sentiment_label", "labels")
if "label" in tokenized_imdb["train"].column_names:
    tokenized_imdb = tokenized_imdb.rename_column("label", "labels")

# Drop None labels using pandas workaround
def drop_none_labels(dataset):
    df = dataset.to_pandas()  # <- FIX: use copy=True explicitly
    df = df[df["labels"].notnull()]
    df["labels"] = df["labels"].astype(int)
    return Dataset.from_pandas(df)

tokenized_goemotions = drop_none_labels(tokenized_goemotions)
tokenized_imdb_train = drop_none_labels(tokenized_imdb["train"])

# Combine datasets
combined_dataset = concatenate_datasets([tokenized_goemotions, tokenized_imdb_train])
combined_dataset = combined_dataset.shuffle(seed=42)

# Keep only necessary columns
combined_dataset = combined_dataset.remove_columns(
    [col for col in combined_dataset.column_names if col not in ["input_ids", "attention_mask", "labels"]]
)

# Convert to TensorFlow-compatible dataset
tokenizer = AutoTokenizer.from_pretrained("cardiffnlp/twitter-roberta-base-sentiment")
data_collator = DataCollatorWithPadding(tokenizer=tokenizer, return_tensors="tf")

tf_dataset = combined_dataset.to_tf_dataset(
    columns=["input_ids", "attention_mask"],
    label_cols="labels",
    shuffle=True,
    batch_size=16,
    collate_fn=data_collator
)

# Preview batch
for batch in tf_dataset.take(1):
    print(batch)
```

# **8 __ Defining and Compiling the Model**

```python
from transformers import TFAutoModelForSequenceClassification
import tensorflow as tf
from tf_keras.optimizers import Adam  # Use tf_keras, NOT tf.keras

# Load the model
model = TFAutoModelForSequenceClassification.from_pretrained(
    "cardiffnlp/twitter-roberta-base-sentiment",
    num_labels=3
)

# Define optimizer from tf_keras
optimizer = Adam(learning_rate=5e-5)

# Compile the model
model.compile(
    optimizer=optimizer,
    loss=tf.keras.losses.SparseCategoricalCrossentropy(from_logits=True),
    metrics=["accuracy"]
)
```

# **9 __ Training the Model**

```python
# Convert to TensorFlow-compatible dataset
tokenizer = AutoTokenizer.from_pretrained("cardiffnlp/twitter-roberta-base-sentiment")
data_collator = DataCollatorWithPadding(tokenizer=tokenizer, return_tensors="tf")

tf_dataset = combined_dataset.to_tf_dataset(
    columns=["input_ids", "attention_mask"],
    label_cols="labels",
    shuffle=True,
    batch_size=16, # The batching is already handled here
    collate_fn=data_collator
)

# Preview batch
for batch in tf_dataset.take(1):
    print(batch)
```

```python
# Train the model
history = model.fit(tf_dataset, epochs=3)

print("\n✅ Training Completed Successfully")
```

# ***10 __ Evaluating the Model***

```python
loss, accuracy = model.evaluate(tf_dataset)
print(f"\n📊 Test Loss: {loss:.4f}")
print(f"✅ Test Accuracy: {accuracy:.4f}")
```

# ***11 __ Making Predictions***

```python
import numpy as np

# Get model predictions
predictions = model.predict(tf_dataset).logits
predicted_labels = np.argmax(predictions, axis=1)

# Extract true labels from the test dataset
true_labels = np.concatenate([y for x, y in tf_dataset], axis=0)

# Print classification report
from sklearn.metrics import classification_report
print(classification_report(true_labels, predicted_labels, target_names=['negative', 'neutral', 'positive']))
```

# ***13 __ Predictions for own Data***

```python
def predict_sentiment(text):
    inputs = tokenizer(text, return_tensors="tf", truncation=True, padding=True, max_length=64)
    outputs = model(inputs)
    probs = tf.nn.softmax(outputs.logits, axis=1)
    prediction = tf.argmax(probs, axis=1).numpy()[0]

    label_map = {0: "negative", 1: "neutral", 2: "positive"}
    confidence = np.max(probs.numpy())

    result = f"Tweet: {text}\n{label_map[prediction].capitalize()} (Confidence: {confidence:.2f})\n"
    return result
```

```python
tweet1 = "The product arrived as expected.",
tweet2 = " nothing  special here .",
tweet3 = " i have neutral feelings for this movie"

print(predict_sentiment(tweet1))
print(predict_sentiment(tweet2))
print(predict_sentiment(tweet3))
```
